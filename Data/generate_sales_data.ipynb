```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ---------------- CONFIG ----------------
num_rows = 50000   # 🔥 BIG DATASET

regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Furniture', 'Clothing']

products = {
    'Electronics': ['Laptop', 'Phone', 'Tablet'],
    'Furniture': ['Chair', 'Table', 'Sofa'],
    'Clothing': ['Shirt', 'Jeans', 'Jacket']
}

base_price = {
    'Laptop': 50000, 'Phone': 20000, 'Tablet': 15000,
    'Chair': 3000, 'Table': 7000, 'Sofa': 20000,
    'Shirt': 1000, 'Jeans': 2000, 'Jacket': 3000
}

# ---------------- GENERATE DATA ----------------
data = []

for i in range(num_rows):
    category = np.random.choice(categories)
    product = np.random.choice(products[category])
    region = np.random.choice(regions)

    # 2 years data
    order_date = pd.to_datetime('2023-01-01') + pd.to_timedelta(
        np.random.randint(0, 730), unit='d'
    )

    quantity = np.random.randint(1, 6)
    price = base_price[product]

    discount = np.random.choice([0, 5, 10, 15, 20])

    # ---- BUSINESS LOGIC ----
    if discount > 15:
        quantity += np.random.randint(1, 3)  # higher discount → more sales

    sales = quantity * price * (1 - discount / 100)

    cost = sales * np.random.uniform(0.6, 0.9)
    profit = sales - cost

    data.append([
        i + 1,
        order_date,
        region,
        category,
        product,
        quantity,
        price,
        discount,
        round(sales, 2),
        round(profit, 2)
    ])

# ---------------- CREATE DATAFRAME ----------------
df = pd.DataFrame(data, columns=[
    'order_id',
    'order_date',
    'region',
    'category',
    'product',
    'quantity',
    'price',
    'discount',
    'sales',
    'profit'
])

# ---------------- SAVE FILES ----------------
df.to_csv("sales_data_big.csv", index=False)
df.to_excel("sales_data_big.xlsx", index=False)

# ---------------- OUTPUT ----------------
print(" BIG Dataset generated successfully!")
print(" Shape:", df.shape)
print("\n Sample Data:")
print(df.head())
```
